```python
import requests
import urllib.parse
import random
from bs4 import BeautifulSoup
import re
import pandas as pd
from datetime import date

titles=[]
links=[]
complaints=[]



for i in range(0,200):
    if i == 0:
        url = 'https://www.sikayetvar.com/sikayetler?brand=teb&viewType=compact'



    else:
        url = 'https://www.sikayetvar.com/sikayetler?brand=teb&viewType=compact'+'&page='+str(i+1)

    print(url)
    r = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36'})
    soup = BeautifulSoup(r.content, 'html.parser')

    results= soup.find_all('div', class_='filter-cards')


    comps = results[0].find_all('article')
    

    for i in range(0,len(comps)):


        compiece=str(comps[i].find_all('a')[0]).split('"')

        
        if '/uye/' in compiece[1]:
            pass
        else:
            
        
           
            titles.append(compiece[3])
            links.append(compiece[1])


            url2 = 'https://www.sikayetvar.com'+compiece[1]

                        
            print(url2)

            r2 = requests.get(url2)
            soup2 = BeautifulSoup(r2.content, 'html.parser')
            complaintpis = (str(soup2).split('p>'))[5]
            
            
            cleanr = re.compile('<.*?>')
            cleancomp = re.sub(cleanr, '', complaintpis)
            cleancomp = cleancomp.split('</')[0]
            cleancomp = cleancomp.replace('\n',' ')

                
                        
            complaints.append(cleancomp)

            dict = {'Basliklar': titles, 'Linkler': links, 'Sikayetler': complaints}  
        
            dataframe = pd.DataFrame(dict)
            dataframe.to_csv('TEB_sikayetvar'+'-'+str(date.today())+'.csv')
```
